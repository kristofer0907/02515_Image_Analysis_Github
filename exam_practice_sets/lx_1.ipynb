```python
from skimage import color, io, measure, img_as_ubyte
import cv2
import numpy as np
import matplotlib.pyplot as plt
dataset = "/home/kkristjansson/DTU/spring2026/02515_Image_Analysis/imageAnalysisGithub/exercises/ex1-IntroductionToImageAnalysis/data/"

im_name = "metacarpals.png"

im_org = io.imread(dataset + im_name)
```

```python
# Lx-1.6: Information about the dimensions of the image and the pixel type

print("Image dimensions:", im_org.shape)
print("Pixel type:", im_org.dtype)

```

```python
# Lx-1.7: Image in grey level scaling and color maps

# Since image is already in gray, we'll directly display it
io.imshow(im_org, cmpap='gray')
io.show()
# To st
io.imshow(im_org, cmap='jet')
io.show()
```

```python
# Lx-1.7: Image Histogram
plt.hist(im_org.ravel(),bins=256) # Convert the image arrays into 1D array and compute the histogram for 256 bins
plt.title("Image histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid()
plt.show()

                                 # and the second is the bin edge


```

```python
# Lx-1.9: Extracting individual bin counts and the bin edges from the histogram
h = plt.hist(im_org.ravel(),bins=256) # h is a list of tuples, where the first element is the bin count
bin_no = 50
count = h[0][bin_no]
print(f"There are {count} values in bin {bin_no}")

bin_left = h[1][bin_no]
bin_right = h[1][bin_no+1]
print(f"Bin edges: {bin_left} to the {bin_right}")
```

```python
# Lx-1.11: Inspect pixel values in an image using row,column indexing
row = 100
col = 150
pixel_value = im_org[row, col]
print(f"Pixel value at row {row} and column {col}: {pixel_value}")
```

```python
# Lx-1.12: Numpy slicing to extract and change pixel values in an image
rows = slice(100, 200) # This will select rows from 100 to 199
cols = slice(150, 250) # This will select columns from 150 to 249
im_org[rows, cols] = 255 # Set the selected region to white
io.imshow(im_org, cmap='gray')
io.show()
```

```python
# Lx-1.13: Computing a binary mask image based on input image
threshold = 0
binary_mask = im_org > threshold
io.imshow(binary_mask, cmap='gray')
io.show()

```

```python
# Lx-1.14: Using a binary mask image to extract and change pixel values in an image
im_org[binary_mask] = 255 # Set all pixels in the original image that correspond
# to True in the binary mask to white
io.imshow(im_org, cmap='gray')
io.show()
```
